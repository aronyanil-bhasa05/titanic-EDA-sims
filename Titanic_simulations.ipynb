```python
import pandas as pd
```

```python
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv("TitanicLogisticReg.csv")
df.sample(10)
```

```python
# y=mx+b , setting up regression where y is dependent var and x is list of input vars

y = df.Survived
x = df[df.columns[2:]]
```

```python
y.sample(10)
```

```python
x.sample(10)
```

```python
# executing log regression

log_reg = sm.Logit(y, x).fit()
```

```python
print(log_reg.summary())
```

```python
log_reg1 = sm.Logit(y, df.Fare).fit()
print(log_reg1.summary())
```

```python
log_reg2 = sm.Logit(y, df[['Fare','Pclass1']]).fit()
print(log_reg2.summary())
```

```python
log_reg3 = sm.Logit(y, df['Pclass1']).fit()
print(log_reg3.summary())
```

```python
log_reg4 = sm.Logit(y, df[['SibSp2','Parch2','Family']]).fit()
print(log_reg4.summary())
```

```python
log_reg5 = sm.Logit(y, df[['SibSp2','Family']]).fit()
print(log_reg5.summary())
```

```python
log_reg_final = sm.Logit(y, df[['Pclass1','SibSp2','Sex1','Embarked1']]).fit()
print(log_reg_final.summary())
```

```python
df_test = pd.read_csv("test_titanic.csv")
```

```python
df_test.sample(10)
```

```python
df_test1 = df_test.copy()
```

```python
# formatting test data according to train data

df_test1['Pclass1'] = np.where(df_test1.Pclass == 1,1,0)
df_test1['Embarked1'] = np.where(~df_test1.Embarked.isin(['Q','S']),1,0)
df_test1['Sex1'] = np.where(df_test1.Sex == 'male',1,0)
df_test1['SibSp2'] = np.where(df_test1.SibSp == 0,0,1)
df_test1 = df_test1[['PassengerId','Pclass1','Sex1','SibSp2','Embarked1']]
df_test2 = df_test1.set_index('PassengerId')
df_test2.sample(10)
```

```python
# Predict the probabilities
predictions = log_reg_final.predict(df_test2)

# Print the predictions
print(predictions)
```

```python
df_pred = pd.DataFrame(predictions, columns=['Survival_prob'])
df_pred1 = df_pred.reset_index(level=0)
df_pred1['Survival_Binary'] = np.where(df_pred1.Survival_prob > 0.55, 1, 0)
df_pred1.sample(10)
```

```python
df_submission = pd.read_csv('gender_submission.csv')
df_submission.sample(10)
```

```python
df_submission1 = df_submission.rename(columns={"Survived": "Survived_Actual"})
df_submission1.sample(10)
```

```python
df_evaluate = pd.merge(df_submission1,df_pred1[['PassengerId', 'Survival_Binary']],on='PassengerId',how='left')
df_evaluate
```

```python
from sklearn.metrics import roc_auc_score, confusion_matrix, f1_score, accuracy_score
```

```python
conf_matrix = confusion_matrix(df_evaluate.Survived_Actual, df_evaluate.Survival_Binary)
conf_matrix
```
